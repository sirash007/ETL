```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

# Hospital File - hospitals by city

```python
#hospital_file = "Resources/indivs_Minnesota18.csv"
hospital_file = "Resources/Minnesota_Hospitals.csv"
hospital_df = pd.read_csv(hospital_file, encoding = 'utf8')
hospital_df.head()
```

# Transform Hospital File

```python
#Step Transform Hospital DF File
# Create a filtered dataframe from specific columns
hospital_cols = ["OBJECTID", "NAME", "ADDRESS", "CITY", "STATE", "HOSP_BEDS"]
hosptial_transformed = hospital_df[hospital_cols].copy()

# Rename the column headers
hosptial_transformed = hosptial_transformed.rename(columns={"OBJECTID": "HOSPITAL_ID",})

# Clean the data by dropping duplicates and setting the index
hosptial_transformed.drop_duplicates("HOSPITAL_ID", inplace=True)
#premise_transformed.set_index("id", inplace=True)
hosptial_transformed.set_index("HOSPITAL_ID", inplace=True)

hosptial_transformed.head()
```

# mndivs file - donations by city

```python
#mndivs_file = "Resources/indivs_Minnesota18.csv"
mndivs_file = "Resources/indivs_Minnesota18.csv"
mndivs_df = pd.read_csv(mndivs_file, encoding = 'ISO-8859-1')
mndivs_df.head()
```

# Transform mndivs file

```python
#Step Transform mndivs DF File
# Create a filtered dataframe from specific columns
mndivs_cols = ["Contribid", "Orgname", "Amount", "City", "State", "Zip", "Gender"]
mndivs_transformed = mndivs_df[mndivs_cols].copy()

# Rename the column headers
mndivs_transformed = mndivs_transformed.rename(columns={"Contribid": "Contribution_ID"})

#mddivs_cols = ["OBJECTID", "NAME", "ADDRESS", "CITY", "STATE", "HOSP_BEDS"]

# Clean the data by dropping duplicates and setting the index
mndivs_transformed.drop_duplicates("Contribution_ID", inplace=True)
#premise_transformed.set_index("id", inplace=True)
mndivs_transformed.set_index("Contribution_ID", inplace=True)
#Validate data type for Amount is Int df.dtypes
#mndivs_transformed.dtypes
mndivs_transformed.head()
```

### Create database connection

```python
connection_string = "postgres:9169@localhost:5432/ETL_Homework"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
hosptial_transformed.to_sql(name='Hospitals', con=engine, if_exists='append', index=True)
```

```python
mndivs_transformed.to_sql(name='Contribution', con=engine, if_exists='append', index=True)
```
